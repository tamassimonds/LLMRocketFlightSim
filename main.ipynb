```python
from rocket_interface import RocketInterface, MotorType, WeatherType

rocket = RocketInterface(name="Calisto")

# Setup the environment
# Default is White Sands, NM with tomorrow noon as the launch date
rocket.setup_environment(
    weather_type=WeatherType.STANDARD  # Using standard atmosphere for simplicity
)
```

```python


# Setup the motor
# Default is Cesaroni M1670
rocket.setup_motor()

# Setup the rocket's basic parameters
rocket.setup_rocket(
    radius=127 / 2000,  # 127mm
    mass=15.0,     # 15kg without motor
    inertia=(6.0, 6.0, 0.25),
    motor_position=0.0  # Motor position from the rocket's origin
)

# Add components to the rocket
rocket.add_nose_cone(
    length=0.55,
    kind="vonKarman",
    position=1.68
)

rocket.add_fins(
    n=4,                # 4 fins
    root_chord=0.120,   # 120mm
    tip_chord=0.040,    # 40mm
    span=0.100,         # 100mm
    position=0.1        # Position from the rocket's origin
)

rocket.add_rail_buttons(
    upper_position=1.2,
    lower_position=0.2
)

# Add a main parachute that deploys at apogee
rocket.add_parachute(
    name="Main",
    cd_s=10.0,
    trigger="apogee",
    lag=1.5    # 1.5 seconds after apogee
)
```

```python
rocket.all_info()
```

```python

flight = rocket.simulate_flight(
    rail_length=5.0,
    inclination=85.0,   # 85 degrees from horizontal
    heading=0.0,        # North
    max_time=5,       # Simulate for up to 5 seconds
    terminate_on_apogee=False
)
```
